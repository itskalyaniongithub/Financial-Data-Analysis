```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# loading csv file to a data frame

```python
df = pd.read_csv('BCG.csv')
```

```python
df.info()
```

```python
df
```

```python
df.head(3)
```

# finding growth percentage for eace attribute over the years

```python
df["Revenue Growth(%)"] = df.groupby(['Company Name'])['Total Revenue'].pct_change()*100
df["Income Growth(%)"] = df.groupby(['Company Name'])['Total income'].pct_change()*100
df["Assets Growth(%)"] = df.groupby(['Company Name'])['Total Assets'].pct_change()*100
df["Liabilities Growth(%)"] = df.groupby(['Company Name'])['Total Liabilities'].pct_change()*100
df["OA Growth(%)"] = df.groupby(['Company Name'])['cash from OA'].pct_change()*100
```

# Filling the null values

```python
df.fillna(df.groupby(['Company Name']).transform('median'), inplace=True)
```

```python
df.head()
```

# Extracting company wise data into company specific data frame

```python
Apple_data = df[df['Company Name'] == 'Apple']
Tesla_data = df[df['Company Name'] == 'Tesla']
Microsoft_data = df[df['Company Name'] == 'Microsoft']
```

# Plotting the growth of different features for different companies for comparision

```python
plt.plot(Apple_data['year'],Apple_data['Revenue Growth(%)'],label="apple")
plt.plot(Tesla_data['year'],Tesla_data['Revenue Growth(%)'],label="tesla")
plt.plot(Microsoft_data['year'],Microsoft_data['Revenue Growth(%)'],label='microsoft')

plt.xlabel('Year')
plt.ylabel('Revenue growth(%)')
plt.title('Revenue growth comparision')
plt.legend()
```

```python
plt.plot(Apple_data['year'],Apple_data['Income Growth(%)'],label="apple")
plt.plot(Tesla_data['year'],Tesla_data['Income Growth(%)'],label="tesla")
plt.plot(Microsoft_data['year'],Microsoft_data['Income Growth(%)'],label='microsoft')

plt.xlabel('Year')
plt.ylabel('Income Growth(%)')
plt.title('Income Growth(%) comparision')
plt.legend()
```
